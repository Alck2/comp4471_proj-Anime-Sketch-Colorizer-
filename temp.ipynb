```python
import cv2
import matplotlib.pyplot as plt
#bgr_img = cv2.imread(r'C:\\Users\\user\\imgs_256' + r'\training'+str(1)+".jpg")
bgr_img = cv2.imread(r".\\dataset_sample\\test.jpg")
cv2.imwrite(r".\\dataset\test\color\test.png", bgr_img)
img_grey = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2GRAY)
img_edge = cv2.adaptiveThreshold(img_grey, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, blockSize=9, C=10)
output = cv2.cvtColor(img_edge, cv2.COLOR_RGB2BGR)
plt.imshow(output)
output = cv2.cvtColor(output, cv2.COLOR_RGB2BGR)
cv2.imwrite(r".\\dataset\reference\temp\test.png", output)
```
